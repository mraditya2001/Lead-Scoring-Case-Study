### Abstract:
The purpose of this project is to develop a logistic regression model to assign a lead 
score ranging from 0 to 100 for each potential lead. The lead score will enable the 
company to effectively target and prioritize leads based on their likelihood of 
conversion. A higher score indicates a "hot" lead, indicating a higher probability of 
conversion, while a lower score suggests a "cold" lead with minimal chances of 
conversion.
### OBJECTIVE:
The purpose of this project is to develop a logistic regression model to assign a lead 
score ranging from 0 to 100 for each potential lead. The lead score will enable the 
company to effectively target and prioritize leads based on their likelihood of 
conversion. A higher score indicates a "hot" lead, indicating a higher probability of 
conversion, while a lower score suggests a "cold" lead with minimal chances of 
conversion.
### INTRODUCTION:
By utilizing logistic regression, we aim to leverage historical lead data and associated 
features to build a predictive model. The model will learn patterns and relationships 
between the lead attributes and conversion outcomes to assign an appropriate lead 
score. Features such as demographic information, past interactions, source of lead,
and other relevant data will be considered as inputs to the model.
### METHODOLOGY:
The project will involve several steps, including data collection, preprocessing, 
exploratory data analysis, feature engineering, model training, and evaluation. The 
logistic regression model will be trained using a suitable optimization algorithm to 
estimate the coefficients and intercept, enabling the calculation of lead scores.
To assess the model's performance and determine an optimal cutoff point for lead 
scores, metrics such as precision, recall, and the receiver operating characteristic 
(ROC) curve will be analysed. These measures will aid in evaluating the model's ability 
to correctly identify hot leads while minimizing false positives.
The successful implementation of this logistic regression model will empower the 
company to allocate its resources effectively and focus on leads with higher conversion 
potential. By prioritizing hot leads, the company can optimize its marketing and sales 
efforts, increasing overall conversion rates and maximizing return on investment.
Overall, this project aims to provide a practical solution for lead scoring, enabling the 
company to make data-driven decisions and improve its lead generation and conversion 
strategies.

### Droping columns with null value greater than 40%

```python
cols = leads.columns
for i in cols:
    if leads[i].isnull().sum()*100/leads.shape[0] > 40:
        leads.drop(i, axis = 1, inplace = True)
```

```python
leads.isnull().sum()*100/leads.shape[0]
```

```python
cols = leads.columns
for i in cols:
    if leads[i].isnull().sum()*100/leads.shape[0] < 15 and leads[i].dtype =='object':
        leads[i].replace(np.nan, leads[i].mode, inplace = True)
```

```python
leads.isnull().sum()*100/leads.shape[0]
```

```python
# Imputing missing value >15 and <40 to "other" category
cols = leads.columns
for i in cols:
    if leads[i].isnull().sum()*100/leads.shape[0] > 15 and leads[i].dtype == 'object':
        leads[i].replace(np.nan, "Others", inplace = True)
```

```python
cat_cols = list(leads.select_dtypes(include='object'))
cat_cols
```

```python
# 'Lead Origin'
a = pd.get_dummies(leads['Lead Origin'], prefix='Lead Origin', drop_first = True, dtype = int)
leads_final0 = pd.concat([leads,a], axis=1)

# 'Lead Source'
b = pd.get_dummies(leads['Lead Source'], prefix='Lead Source', drop_first = True, dtype = int)
leads_final1 = pd.concat([leads_final0,a], axis=1)

# 'Last Activity'
c = pd.get_dummies(leads['Last Activity'], prefix='Last Activity', drop_first = True, dtype = int)
leads_final2 = pd.concat([leads_final1,a], axis=1)

# 'Country'
d = pd.get_dummies(leads['Country'], prefix='Country', drop_first = True, dtype = int)
leads_final3 = pd.concat([leads_final2,a], axis=1)

# 'Specialization'
e = pd.get_dummies(leads['Specialization'], prefix='Specialization', drop_first = True, dtype = int)
leads_final4 = pd.concat([leads_final3,a], axis=1)

# 'What is your current occupation'
f = pd.get_dummies(leads['What is your current occupation'], prefix='What is your current occupation', drop_first = True, dtype = int)
leads_final5 = pd.concat([leads_final4,a], axis=1)

# 'What matters most to you in choosing a course'
g = pd.get_dummies(leads['What matters most to you in choosing a course'], prefix='What matters most to you in choosing a course', drop_first = True, dtype = int)
leads_final6 = pd.concat([leads_final5,a], axis=1)

# 'Tags'
h = pd.get_dummies(leads['Tags'], prefix='Tags', drop_first = True, dtype = int)
leads_final7 = pd.concat([leads_final6,a], axis=1)

# 'City'
i = pd.get_dummies(leads['City'], prefix='City', drop_first = True, dtype = int)
leads_final8 = pd.concat([leads_final7,a], axis=1)

# 'Last Notable Activity'
j = pd.get_dummies(leads['Last Notable Activity'], prefix='Last Notable Activity', drop_first = True, dtype = int)
leads_final = pd.concat([leads_final8,a], axis=1)
```
